# Aliasing

```python
import numpy as np
import matplotlib.pyplot as plt

# Original signal parameters
f_signal = 20  # Frequency of the cosine signal (Hz)
t = np.linspace(0, 1, 1000, endpoint=False)  # Continuous time vector
signal = np.cos(2 * np.pi * f_signal * t)  # Original continuous signal

# Sampling parameters
f_sample = 25  # Sampling frequency (Hz)
t_sampled = np.arange(0, 1, 1 / f_sample)  # Sampled time points
samples = np.cos(2 * np.pi * f_signal * t_sampled)  # Sampled values

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(t, signal, label='Original Signal', alpha=0.7)  # Plot original signal
plt.stem(t_sampled, samples, linefmt='r-', markerfmt='ro', basefmt=" ", label='Sampled Signal')
plt.title('Aliasing Demonstration (20 Hz Cosine, Sampled at 25 Hz)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.show()
```

# Signal Reconstruction

```python
from scipy.signal import resample

# Reconstructing the signal using high sampling rate
num_samples = 1000
reconstructed_signal = resample(samples, num_samples)

# Plotting the reconstruction
plt.figure(figsize=(10, 6))
plt.plot(t, signal, label='Original Signal')
plt.plot(t, reconstructed_signal, label='Reconstructed Signal', linestyle='--')
plt.title('Signal Reconstruction')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.show()
```

# Cosine Wave

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
f = 20  # Frequency (Hz)
t = np.linspace(0, 1, 1000, endpoint=False)  # Time vector
cos_wave = np.cos(2 * np.pi * f * t)

# Plotting
plt.plot(t, cos_wave, label="Cose Wave")
plt.title("Cose Wave")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.legend()
plt.show()
```

# Delta Encoding and decoding

```python
import numpy as np

# Delta Encoding
def delta_encode(signal):
    return np.diff(np.insert(signal, 0, signal[0]))

# Delta Decoding
def delta_decode(encoded_signal):
    return np.cumsum(encoded_signal)

# Example
original_signal = [8, 10, 12, 12, 14]
encoded_signal = delta_encode(original_signal)
decoded_signal = delta_decode(encoded_signal)

print("Original Signal:", original_signal)
print("Encoded Signal:", encoded_signal)
print("Decoded Signal:", decoded_signal)
```
