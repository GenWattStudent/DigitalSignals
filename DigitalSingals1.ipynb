# Analiza Wektorów i IDFT
Ten notebook zawiera analizę zadanych wektorów oraz syntezę sygnału za pomocą IDFT.

Tworzenie wykresu (część rzeczywista, część urojona).

```python
import numpy as np
import matplotlib.pyplot as plt

x_mu = np.array([10, 5, 6, 6, 2, 4, 3, 4, 5, 0, 0, 0, 0], dtype=complex)
N = len(x_mu)

k = np.arange(N).reshape(N, 1)
n = np.arange(N).reshape(1, N)
W = np.exp(2j * np.pi * k @ n / N) / N 

time_signal = W @ x_mu

plt.figure(figsize=(10, 5))

plt.stem(np.real(time_signal), markerfmt='C0o', basefmt='C0:', linefmt='C0:', label='Real Part')
plt.stem(np.imag(time_signal), markerfmt='C1o', basefmt='C1:', linefmt='C1:', label='Imaginary Part')

plt.plot(np.real(time_signal), 'C0-', lw=0.5)
plt.plot(np.imag(time_signal), 'C1-', lw=0.5)

plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('Synthesized Signal using IDFT')

plt.legend()
plt.grid()
plt.show()
```
